```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_csv = pd.read_csv('Extrato.csv',delimiter = ';',header=3)
```

```python
df_csv.head()
```

```python
df_csv.tail()
```

```python
transacoes_por_data = df_csv['Data Lançamento'].value_counts().sort_index()

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
transacoes_por_data.plot(kind='bar', color='skyblue')
plt.title('Quantidade de Transações por Data de Lançamento')
plt.xlabel('Data de Lançamento')
plt.ylabel('Quantidade de Transações')
plt.xticks(rotation=45)
plt.show()
```

```python
def float_convert(valor):
    valor = float(valor.replace(',','.'))
    return valor

df_csv['Valor'] = df_csv['Valor'].apply(float_convert)


df_csv
```

```python
## Maior gasto por descrição

gastos_por_descricao = df_csv.groupby('Descrição')['Valor'].sum().sort_values()

gastos_por_descricao
```

```python
gastos_por_Data = df_csv.groupby('Data Lançamento')['Valor'].sum()

gastos_por_Data
```

```python

##df_csv['Data Lançamento'] = pd.to_datetime(df_csv['Data Lançamento'],format='%d/%m/%Y')
gastos_por_data = df_csv.groupby('Data Lançamento')['Valor'].sum()

# Criando o gráfico de barras
plt.figure(figsize=(12, 6))
gastos_por_data.plot(kind='bar', color='salmon')
plt.title('Valores Gastos por Data de Lançamento')
plt.xlabel('Data de Lançamento')
plt.ylabel('Valor Gasto (R$)')
plt.xticks(rotation=45)
plt.show()
```
